```python
#STEP 1:
#Load the dataset
import pandas as pd
df=pd.read_csv("insurance_policy_dataset.csv")
```

```python
df
```

```python
#Step 2:
df.drop(columns=["policy_ID"],axis=1,inplace=True)
df.head(5)
```

```python
#step3:Text preprocessing
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("punkt_tab")
```

```python
#step4:
stop_words=set(stopwords.words("english"))
lemma=WordNetLemmatizer()
import re
def clean(text):
  if text is None:
    return ''
  text=re.sub(r'https\S+','',text)
  text=re.sub('[^a-zA-Z]'," ",text)
  text=text.lower()
  text=word_tokenize(text)
  text=[lemma.lemmatize(word=w,pos='v')for w in text if w not in stop_words if len(w)>2]
  text=' '.join(text)
  return text
df['cleanedtext']=df['Policy_Text_EN'].apply(clean)
df['cleanedsummary']=df['Summarized_Text_EN'].apply(clean)
```

```python
df.drop(columns=['Policy_Text_EN','Summarized_Text_EN'],axis=1,inplace=True)
df.head(5)
```

```python
df1 = df
df1.to_csv("cleaned_insurance_text.csv")
print("‚úÖ CSV file saved as 'cleaned_insurance_text.csv'")
```

```python
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Load dataset
df2 = pd.read_csv("cleaned_insurance_text.csv")
print("‚úÖ CSV loaded")

texts = df2["cleanedtext"].astype(str).tolist()

# Load model & tokenizer (CPU)
model_name = "facebook/nllb-200-distilled-600M"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
model.eval()

print("üß† Model loaded on CPU")

# Language codes
LANG_CODES = {
    "tamil": "tam_Taml",
    "hindi": "hin_Deva",
    "telugu": "tel_Telu",
    "kannada": "kan_Knda"
}

# Batch translation function (CPU-friendly)
def translate_batch(texts, target_lang_code):
    tokenizer.src_lang = "eng_Latn"

    inputs = tokenizer(
        texts,
        return_tensors="pt",
        padding=True,
        truncation=True,
        max_length=256
    )

    with torch.inference_mode():
        outputs = model.generate(
            **inputs,
            forced_bos_token_id=tokenizer.convert_tokens_to_ids(target_lang_code),
            max_length=256
        )

    return tokenizer.batch_decode(outputs, skip_special_tokens=True)

# Translate once per language (only 4 model calls total)
for lang, code in LANG_CODES.items():
    print(f"üåç Translating to {lang}...")
    df2[f"cleanedtext_{lang}"] = translate_batch(texts, code)

# Save results
df2.to_csv("translated_insurance_text.csv", index=False)
print("‚úÖ Translations saved to 'translated_insurance_text.csv'")
```

```python
import streamlit as st
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

st.set_page_config(page_title="Insurance Translator", layout="centered")
st.title("üåç Insurance Text Translator")

LANG_CODES = {
    "Tamil": "tam_Taml",
    "Hindi": "hin_Deva",
    "Telugu": "tel_Telu",
    "Kannada": "kan_Knda"
}

MODEL_PATH = "nllb_insurance_translator"

@st.cache_resource
def load_model():
    tokenizer = AutoTokenizer.from_pretrained(MODEL_PATH)
    model = AutoModelForSeq2SeqLM.from_pretrained(MODEL_PATH)
    model.eval()
    return tokenizer, model

tokenizer, model = load_model()

def translate_text(text, target_lang_code):
    tokenizer.src_lang = "eng_Latn"

    inputs = tokenizer(
        [text],
        return_tensors="pt",
        truncation=True,
        padding=True,
        max_length=256
    )

    with torch.inference_mode():
        output = model.generate(
            **inputs,
            forced_bos_token_id=tokenizer.convert_tokens_to_ids(target_lang_code),
            max_length=256,
            num_beams=1
        )

    return tokenizer.decode(output[0], skip_special_tokens=True)

# UI
text_input = st.text_area("Enter insurance text (English):")

lang = st.selectbox("Select target language", list(LANG_CODES.keys()))

if st.button("Translate"):
    if text_input.strip():
        with st.spinner("Translating..."):
            translated = translate_text(text_input, LANG_CODES[lang])
        st.success("Translation complete")
        st.text_area("Translated Text", translated, height=150)
    else:
        st.warning("Please enter text")
```
